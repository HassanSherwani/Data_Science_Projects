# Hierarchical clustering

# Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
X = dataset.iloc[:, [8, 12]].values
```

```python
type(X)
```

# Using the dendogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

**A vertical line without being intersected will show our optimal clusters. We can see that red line on left end shows that trend. So we will make a horizontal line and see how many points would it cut.<br>
It cuts 4 vertical lines so our clusters are 4

# Fitting hierarchical clustering to the mall dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

# Visualizing the clusters

```python
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], s=100, c='cyan', label='Cluster 4')


plt.title('Clusters of clients')
plt.xlabel('Balance')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
